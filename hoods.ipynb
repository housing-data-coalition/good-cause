```python
import pandas as pd
import geopandas as gpd
from shapely.geometry import Point
```

```python
neighborhoods_gdf = gpd.read_file('2010 Neighborhood Tabulation Areas (NTAs).geojson')
```

```python
gce_df = pd.read_csv('data/likely-gce-bbls_2024-06-24.csv').drop_duplicates()
```

```python
gce_df
```

```python
gce_df['geometry'] = gce_df.apply(
    lambda r: Point(r['longitude'], r['latitude']), 1
)
```

```python
gce_gdf = gpd.GeoDataFrame(gce_df[['bbl', 'longitude', 'latitude']], geometry=gce_df['geometry'], crs="EPSG:4326")
```

```python
gce_gdf
```

```python
points_in_neighborhoods = gpd.sjoin(gce_gdf, neighborhoods_gdf, how='left', predicate='within')
```

```python
points_in_neighborhoods
```

```python
points_in_neighborhoods[['bbl', 'longitude', 'latitude', 'ntaname', 'ntacode', 'county_fips']].to_csv("neighborhood_gce_df.csv", index=False)
```
